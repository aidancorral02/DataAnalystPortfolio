### Analytical on Categorical and Discrete Feature

```python
# Create new Dataframe for Categorical Features
categorical = []
for column in data.columns:
    if data[column].dtype == object:
        categorical.append(column)
        print(f'{column} : {data[column].unique()}')
        print('====================================')
```

```python
categorical
```

```python
# Make a list of categorical feature and use sub plot to plot multiple feature
categorical_col = data[['Gender','EducationBackground',
                        'MaritalStatus',
                        'BusinessTravelFrequency',
                        'OverTime','Attrition']]
# defining canvas size
plt.figure(figsize=(20,22)) 
# counter
plotno = 1 

# iteration of columns / acessing the columns from count 
for column in categorical_col: 
    # set the limit
    if plotno<=7:
        # # plotting 6 graphs (3-rows,2-columns) ,plotnumber is for count
        plt.subplot(3,2,plotno)
        # Plotting count plots because the feature is categorical
        sns.countplot(x=categorical_col[column],hue=data.PerformanceRating) 
        # assigning name to x-axis and font size is 20
        plt.xlabel(column,fontsize=20)  
        plt.ylabel('PerformanceRating')
    # counter increment
    plotno+=1 
plt.tight_layout()
# used to hide the storage loction
plt.show() 
```
